```python
# Giving the details of the process using the gradient descent method int the notebook.
# y = 0.5 * (x-0.25) ** 2
# inital value:
# η=0.500
# ε=1e-10



```

```python
import numpy as np
import matplotlib.pyplot as plt

# 定义函数
def f(x):
    return 0.5 * (x - 0.25) ** 2

# 梯度函数
def grad_f(x):
    return x - 0.25

# 梯度下降函数
def gradient_descent(x, eta, eps):
    xs = [x]
    while True:
        grad = grad_f(x)
        x -= eta * grad
        xs.append(x)
        if abs(grad) < eps:
            break
    return xs

# 初始化参数
x0 = 4
eta = 0.5
eps = 1e-10

# 进行梯度下降
xs = gradient_descent(x0, eta, eps)

# 输出每次迭代的结果
for i, x in enumerate(xs):
    print(f"Iteration {i}: x = {x:.8f}, y = {f(x):.8f}")

# 绘制梯度下降过程
fig, ax = plt.subplots()
ax.plot(np.arange(-1, 3, 0.01), f(np.arange(-1, 3, 0.01)), 'k-', label="y=0.5*(x-0.25)^2")
ax.plot(xs, f(np.array(xs)), 'bo-', label="Gradient Descent")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.legend()
plt.show()
```
