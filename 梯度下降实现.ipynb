```python
import matplotlib.pyplot as plt
import numpy as np

# Define the function y = x^2
def f(x):
    return x**2

# Define the derivative of the function y = x^2
def df(x):
    return 2*x

# Set the initial value of x to -4
x = -4

# Set the learning rate
learning_rate = 0.9

# Set the threshold for convergence
threshold = 0.01

# Initialize the change in x to a large value
dx = float('inf')

# Create lists to store the values of x and f(x) at each iteration
x_values = [x]
y_values = [f(x)]

# Perform gradient descent
while abs(dx) >= threshold:
    # Compute the gradient
    gradient = df(x)
    
    # Update x and compute the change in x
    x_new = x - learning_rate * gradient
    dx = x_new - x
    x = x_new
    
    # Append the current value of x and f(x) to the lists
    x_values.append(x)
    y_values.append(f(x))

# Define the range of x (-8, 12) and set the interval to 1
x = np.arange(-8, 20)
y = f(x)

# Create a figure and specify its size
plt.figure(figsize=(8, 5))

# Plot the graph of y=x^2, set color to red, line width to 1, and line style to --
plt.plot(x, y, color='red', linewidth=1.0, linestyle='--')

# Plot the gradient descent process, set color to blue and marker to o
plt.plot(x_values, y_values, color='blue', marker='o')

# Set the range and labels for the x and y axes
plt.xlim(-8, 12)
plt.ylim(0, 150)
plt.xlabel('x')
plt.ylabel('y')

# Show the plot
plt.show()
```
